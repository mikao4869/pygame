```python
import pygame

# 초기화
pygame.init()
width, height = 800, 400
WHITE = (255, 255, 255)
screen = pygame.display.set_mode((width, height))
pygame.display.set_caption("공룡 게임")
clock = pygame.time.Clock()
font = pygame.font.SysFont(None, 30)  # 기본 폰트

# 공룡
dino = pygame.Rect(50, 300, 50, 50)
is_jumping = False
jump_point = 0
gravity = 1

# 장애물
obstacle = pygame.Rect(width + 20, 300, 50, 50)
obstacle_speed = 10

obstacle2 = pygame.Rect(width + 500, 300, 50, 50)  # 시작부터 떨어뜨려 놓음
obstacle2_speed = 10

# 점수
point = 0

# 최소 간격
min_gap = 250

# 게임 루프
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        if event.type == pygame.KEYDOWN and event.key == pygame.K_SPACE and not is_jumping:
            jump_point = -20
            is_jumping = True

    # 점프
    dino.y += jump_point
    jump_point += gravity
    if dino.y >= 300:
        dino.y = 300
        is_jumping = False

    # 장애물 이동
    obstacle.x -= obstacle_speed
    obstacle2.x -= obstacle2_speed

    # obstacle 재배치
    if obstacle.x < -50:
        # obstacle2와 간격 유지
        new_x = max(obstacle2.x + min_gap, width + 20)
        obstacle.x = new_x

    # obstacle2 재배치
    if obstacle2.x < -50:
        # obstacle과 간격 유지
        new_x = max(obstacle.x + min_gap, width + 20)
        obstacle2.x = new_x

    # 충돌
    if dino.colliderect(obstacle) or dino.colliderect(obstacle2):
        print("게임종료")
        running = False

    # 점수
    point += 0.08
    point_text = font.render(f"{int(point)}", True, (0, 0, 0))

    # 화면 그리기
    screen.fill(WHITE)
    pygame.draw.rect(screen, (255, 0, 0), dino)
    pygame.draw.rect(screen, (0, 0, 255), obstacle)
    pygame.draw.rect(screen, (255, 255, 0), obstacle2)
    screen.blit(point_text, (600, 20))

    pygame.display.update()
    clock.tick(60)

pygame.quit()
```
