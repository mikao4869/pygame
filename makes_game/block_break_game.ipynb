```python
import pygame
import sys

pygame.init()
screen = pygame.display.set_mode((600, 800))
pygame.display.set_caption("벽돌 깨기")
clock = pygame.time.Clock()

# 폰트 설정
font = pygame.font.SysFont(None, 48)

# 공(원형)
ball_radius = 10
ball_x = 310  # 중심좌표 (초기값)
ball_y = 410
ball_speed = [5, -5]

# 발판
floor = pygame.Rect(250, 650, 100, 10)

# 블럭
bricks = []
bricks_color = (0, 255, 0)
for row in range(5):
    for col in range(10):
        brick = pygame.Rect(60 + col * 50, 40 + row * 30, 40, 20)
        bricks.append(brick)

score = 0  # 점수 변수

# 충돌 판정 함수
def circle_collision(circle_x, circle_y, radius, rect):
    close_x = max(rect.left, min(circle_x, rect.right))
    close_y = max(rect.top, min(circle_y, rect.bottom))
    direct_x = circle_x - close_x
    direct_y = circle_y - close_y
    return (direct_x**2 + direct_y**2) <= radius**2

running = True

while running:
    screen.fill((0, 0, 0))

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # 패들의 움직임 구현
    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT] and floor.left > 0:
        floor.x -= 7
    if keys[pygame.K_RIGHT] and floor.right < 600:
        floor.x += 7

    # 공 움직이기
    ball_x += ball_speed[0]
    ball_y += ball_speed[1]

    # 벽에 부딪히면 공의 방향을 바꾸기
    if ball_x - ball_radius <= 0 or ball_x + ball_radius >= 600:
        ball_speed[0] *= -1
    if ball_y - ball_radius <= 0:
        ball_speed[1] *= -1

    # 발판과 부딪히면 공의 방향을 바꿈
    if circle_collision(ball_x, ball_y, ball_radius, floor):
        ball_speed[1] *= -1

    # 벽돌과 부딪혔을 때 벽돌을 사라지게 만듦
    hit_brick = None
    for idx, brick in enumerate(bricks):
        if circle_collision(ball_x, ball_y, ball_radius, brick):
            hit_brick = idx
            break
    if hit_brick is not None:
        bricks.pop(hit_brick)
        ball_speed[1] *= -1
        score += 10  # 점수 10점 추가

    # 공이 바닥에 닿으면 게임 종료
    if ball_y + ball_radius >= 800:
        running = False
        print(f"점수: {score}")

    # 그리기
    pygame.draw.rect(screen, (255, 255, 255), floor)
    pygame.draw.circle(screen, (255, 0, 0), (ball_x, ball_y), ball_radius)  # 공
    for brick in bricks:
        pygame.draw.rect(screen, bricks_color, brick)

    # 점수 표시
    score_text = font.render(f"Score: {score}", True, (255, 255, 0))
    screen.blit(score_text, (20, 10))

    pygame.display.update()
    clock.tick(60)

pygame.quit()
sys.exit()
```
